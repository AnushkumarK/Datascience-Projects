```python
#importing files

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

appData=pd.read_csv(r'C:\Users\91865\Downloads\Data Science\Case study\previous_application.csv',encoding = 'unicode_escape')
appData1=pd.read_csv(r'C:\Users\91865\Downloads\Data Science\Case study\application_data.csv',encoding = 'unicode_escape')
#print (appData)


```

```python
appData.describe()
```

```python
#Creating a subset of relevant columns for analysis

newset=appData.drop(['SK_ID_PREV','NAME_CASH_LOAN_PURPOSE','WEEKDAY_APPR_PROCESS_START', 'CODE_REJECT_REASON','HOUR_APPR_PROCESS_START','FLAG_LAST_APPL_PER_CONTRACT', 'NFLAG_LAST_APPL_IN_DAY','NAME_TYPE_SUITE','RATE_DOWN_PAYMENT', 'RATE_INTEREST_PRIMARY', 'CHANNEL_TYPE', 'SELLERPLACE_AREA', 'NAME_SELLER_INDUSTRY', 'NAME_YIELD_GROUP', 'PRODUCT_COMBINATION','DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE', 'DAYS_LAST_DUE_1ST_VERSION','DAYS_LAST_DUE', 'DAYS_TERMINATION', 'NFLAG_INSURED_ON_APPROVAL','RATE_INTEREST_PRIVILEGED','DAYS_DECISION', 'NAME_PAYMENT_TYPE'], axis=1)
print(newset.columns)
```

```python
#Finding percentage of missing values

missingvalues=newset.isnull().sum()*100/len(newset)
missingvaluessorted=pd.DataFrame({'column_name':newset.columns,
                                               'missingvalues': missingvalues})
missingvaluessorted.sort_values('missingvalues', inplace=True,ascending=False)
missingvaluessorted
```

```python
#Imputing missing values

newset=newset.apply(lambda x:x.fillna(x.value_counts().index[0]))
newset.head()
percent_missingvalues=newset.isnull().sum()*100/len(newset)
percent_missingvaluessorted=pd.DataFrame({'column_name':newset.columns,
                                               'percent_missingvalues':percent_missingvalues})
percent_missingvaluessorted.sort_values('percent_missingvalues', inplace=True,ascending=False)
percent_missingvaluessorted
```

```python
#dropping rows which have XNA values

newset=newset[~newset.NAME_CONTRACT_STATUS.str.contains("XNA")]
newset=newset[~newset.NAME_CLIENT_TYPE.str.contains("XNA")]
newset=newset[~newset.NAME_PORTFOLIO.str.contains("XNA")]
newset=newset[~newset.NAME_GOODS_CATEGORY.str.contains("XNA")]
newset=newset[~newset.NAME_PRODUCT_TYPE.str.contains("XNA")]
newset
```

```python
newset.info()
```

```python
#detecting outliers
plt.figure(figsize=(10,10))
np.random.seed()
boxplot = newset.boxplot(column=['AMT_ANNUITY', 'AMT_APPLICATION', 'AMT_CREDIT','AMT_DOWN_PAYMENT','AMT_GOODS_PRICE','CNT_PAYMENT'])
```

```python
#removing outliers

from scipy import stats
plt.figure(figsize=(10,10))
newset1=newset.drop(['NAME_CONTRACT_TYPE','NAME_CONTRACT_STATUS','NAME_CLIENT_TYPE','NAME_GOODS_CATEGORY','NAME_PORTFOLIO','NAME_PRODUCT_TYPE','CNT_PAYMENT'], axis=1).set_index("SK_ID_CURR", inplace=False)
newset2=newset1[(np.abs(stats.zscore(newset1)) < 3).all(axis=1)]
boxplot_new=newset2.boxplot()
```

```python
# Including 'TARGET' column from application_data set in previous_application data

miniAppData = appData1[["SK_ID_CURR","TARGET"]]
mergedData = pd.merge(newset, miniAppData, on='SK_ID_CURR', how='inner')
#print (mergedData.columns)
mergedData.info()
```

```python
#Data Imbalance and checking data categories

freq = mergedData['TARGET'].value_counts()
freq1 = mergedData['NAME_CONTRACT_TYPE'].value_counts()
freq2= mergedData['NAME_CONTRACT_STATUS'].value_counts()
freq3 = mergedData['NAME_CLIENT_TYPE'].value_counts()
freq4 = mergedData['NAME_GOODS_CATEGORY'].value_counts()
freq5 = mergedData['NAME_PORTFOLIO'].value_counts()
freq6 = mergedData['NAME_PRODUCT_TYPE'].value_counts()
print(freq,'\n',freq1,'\n',freq2,'\n',freq3,'\n',freq4,'\n',freq5,'\n',freq6)
```

```python
#Splitting the dataframes into defaulters(def) and non defaulters(notdef)

mergedData_notdef=mergedData[mergedData['TARGET']==0]
mergedData_notdef.info()
mergedData_def=mergedData[mergedData['TARGET']==1]
mergedData_def.info()
```
